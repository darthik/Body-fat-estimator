```python
def distplot(feature):
    plt.figure(figsize=(10,5))
    sns.distplot(df['BodyFat'],color='magenta',hist=False,label='BodyFat')
    sns.distplot(df[feature],color='red',hist=False,label=feature)
    plt.legend()
    plt.show()
    
cols = list(df.columns)
for feature in cols:
    print(f'Distribution plots for {feature} is below')
    distplot(feature)
    print('-'*100)   
```

```python
def drawplots(df,col):
    plt.figure(figsize=(10,5))
    plt.subplot(1,3,1)
    plt.hist(df[col])
    
    plt.subplot(1,3,2)
    stats.probplot(df[col],dist='norm',plot=plt)
    
    plt.subplot(1,3,3)
    sns.boxplot(df[col])
    
    plt.show()
    
    
cols=list(df.columns)
for col in cols:
    print(f'Distribution plots for the {col} is below')
    drawplots(df,col)
    print('-'*100)
```

```python
def correlation(df,threshold):
    colcor = set()
    cormat = df.corr()
    
    for i in range(len(cormat)):
        for j in range(i):
            
            if abs(cormat.iloc[i][j])>threshold:
                colname = cormat.columns[i]
                colcor.add(colname)
                
    return colcor

cols3 = correlation(features,0.85)
cols3
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn import metrics
```

```python
X_train, X_test,y_train,y_test = train_test_split(totrain, target, test_size=0.2,random_state=20)
X_train.shape,X_test.shape
```

```python
DT_reg = DecisionTreeRegressor(random_state=20)
DT_reg.fit(X_train,y_train)
plt.figure(figsize=(15,7))
tree.plot_tree(DT_reg,filled=True)
```

### The tree is very large and hence we can prune the tree, # We will prune the tree using cost-complexity-pruning-path

```python
path = DT_reg.cost_complexity_pruning_path(X_train,y_train)
ccp_alpha = path.ccp_alphas
```

```python
ccp_alpha.shape
```

```python
ccp_alpha
```

```python
alphalist = []
for i in range(len(ccp_alpha)):
    DT_reg = DecisionTreeRegressor(ccp_alpha=ccp_alpha[i])
    DT_reg.fit(X_train,y_train)
    alphalist.append(DT_reg)
```

```python
len(alphalist)
```

```python
train_score_DT_reg = [alphalist[i].score(X_train,y_train) for i in range(len(alphalist))]

test_score_DT_reg = [alphalist[i].score(X_test,y_test) for i in range(len(alphalist))]
```

```python
plt.figure(figsize=(15,7))
plt.xlabel('Alpha')
plt.ylabel('Accuracy')
plt.plot(ccp_alpha,train_score_DT_reg,label='Training',marker='*',color='magenta',ds='steps')
plt.plot(ccp_alpha,test_score_DT_reg,label='Test',marker='o',color='green',ds='steps')
plt.grid()
plt.legend()
plt.show()
```

##### So the alpha value which has good accuracy is some where between 0 and 1 after that the accuracy keeps declining 

```python
DT_reg_apha = DecisionTreeRegressor(ccp_alpha=0.05)
DT_reg_apha.fit(X_train,y_train)
y_pred = DT_reg_apha.predict(X_test)
print(f'Decision Tree regression R2 score :{metrics.r2_score(y_test,y_pred)}')

RF_reg_alpha = RandomForestRegressor(n_estimators=1000,ccp_alpha=0.05)
RF_reg_alpha.fit(X_train,y_train)
y_pred_RF_reg = RF_reg_alpha.predict(X_test)
print(f'Random Forest regression R2 score: {metrics.r2_score(y_test,y_pred_RF_reg)}')
```

### Hyperparameter tuning

```python
params = {
    'DecisionTree':{
        'model':DecisionTreeRegressor(),
        'params':{
            'criterion':['mse','mae'],
            'splitter':['best','random'],
            'min_samples_split':[1,2,5,10,12],
            'min_samples_leaf':[1,2,5,10,12],
            'max_features':['auto','sqrt'],
            'ccp_alpha':[1, 2,2.5,3,3.5,4,5]
            
            
        }
    },
    'RandomForest':{
        'model':RandomForestRegressor(),
        'params':{
            'n_estimators':[int(x) for x in np.linspace(start=1,stop=1200,num=10)],
            'criterion':['mse','mae'],
            'max_depth':[int(x) for x in np.linspace(start=1, stop=30,num=5)],
            'min_samples_split':[2,5,10,12],
            'min_samples_leaf':[2,5,10,12],
            'max_features':['auto','sqrt'],
            'ccp_alpha':[1,2,2.5,3,3.5,4,5],
        }
    },
    'SVM':{
        'model':SVR(),
        'params':{
            'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
            'tol':[1e-10,1e-5,1e-4,0.025,0.5,0.75],
            'C':[0.25,0.5,0.75,1],
            'max_iter':[int(x) for x in np.linspace(start=1,stop=250,num=10)],
            
        }
    }
}
```

```python
params.items()
```

```python
from sklearn.model_selection import RandomizedSearchCV
scores = []
for modelname,mp in params.items():
    clf = RandomizedSearchCV(estimator=mp['model'],param_distributions=mp['params'],n_iter=10,n_jobs=-1,
                             cv=5,scoring='neg_mean_squared_error')
    clf.fit(X_train,y_train)
    scores.append({
        'Model_name': modelname,
        'Best_score':clf.best_score_,
        'Best_estimator':clf.best_estimator_
    })
                        
```

```python
scores
```

```python
scoresdf =pd.DataFrame(scores,columns=['Model_name', 'Best_score','Best_estimator'])
scoresdf
```
